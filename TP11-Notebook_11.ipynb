#### 1) TRANSFORMACION DE VARIABLES

- La mayor√≠a de los algoritmos de machine learning no admiten `strings` como variables y requieren que las variables sean numericas. Por ese motivo, es necesario convertir las variables categoricas en su representaci√≥n numerica. Para esto hay varias opciones que iremos viendo.

**`LabelEncoder()`**

```python
from sklearn.preprocessing import LabelEncoder# Utilizamos Laber Encoder para cambiar los datos de una celda como En Cuidad que los cambie por 0,1,2,3,etc
```

```python
test_encoder = LabelEncoder() #Ejecutamos la funci√≥n
```

```python
blackfriday.loc[:, 'City_Category'] = test_encoder.fit_transform(blackfriday['City_Category']) #Realizamos el cambio de la categoria
```

```python
blackfriday.head(10) #mostramos las 10 primeras filas con el cambio
```

**`get_dummies()`**

```python
pd.get_dummies(blackfriday, columns=["Gender"]) #Fumci√≥m "DUMIES" lo que hacemos es aperturar m√°s la celda GENDER o Genero en Masculino y Femenino y en la celda que es correcta coloca True sino ser√° False
```

```python
pd.get_dummies(blackfriday["Gender"]) # solo muestro la columna
```

```python
blackfriday[['female', 'male']] = pd.get_dummies(blackfriday["Gender"]) #realizamos el cambio de Gender por las palabra Fenemino y Masculino
```

```python
blackfriday.head() #Muestras las 5 filas con el cambio anterior
```

**`OneHotEncoder()`**

```python
from sklearn.preprocessing import OneHotEncoder #Importamos las variable OneHOtEncoder
```

```python
blackfriday['Gender'].unique() # Vemos los valores unicos
```

```python
gender_encoder = OneHotEncoder()
```

```python
gender_encoder.fit_transform(blackfriday[['Gender']]).toarray() #Lo transforma los valores con el FIT el array que genere con los valores unicos donde el primer valor es Femenino y segundo es masculuno y se puede generar (1,0) o (0,1)
```

```python
gender_encoder.categories_ # Que me muestre las categorias
```

```python
niveles = gender_encoder.categories_[0].tolist() #Ac√° transformo las categorias en una lista para generar la matriz con las columnas y filas
```

```python
one_hot_gender = pd.DataFrame(gender_encoder.fit_transform(blackfriday[['Gender']]).toarray(), columns=niveles)
```

```python
one_hot_gender #Me imprime la matriz
```

```python
new_df = pd.concat([blackfriday, one_hot_gender], axis=1) # Aca lo que realizo es guardarlo en otro DataFrem y lo concateno  con la el data set principal el one_hot_gener que acabamos de crear y sirve para categorizar nuestras variables
```

```python
new_df.head() #Acaba vemos al final las dos columnas creadas y opodemos utilizar esto para clasificar nuestros datos en grupos
```

**`binning`**

```python
print(blackfriday.Age.min()) #vemos el minimo
print(blackfriday.Age.max()) #vemos el maximo
```

```python
blackfriday["Age"] = pd.to_numeric(blackfriday["Age"], errors='coerce')
```

```python
bin_age = [10, 17, 70, 80] # ac√° utilizamos BIN la funci√≥n agrupa los datos para reducir el ruido que produce la gran variedad de datos en un featur (Edad)
labels = ["Adolescente", "Adulto", "Anciano"] #Ac√°los label deben ser uno menos que los bin sino no los vamos a poder ejecutar 3 label y 4 en bins

age_categories = pd.cut(blackfriday["Age"], bins=bin_age, labels=labels)
```
